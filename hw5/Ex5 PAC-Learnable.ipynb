# Instance space

In this exercise, our dataset consists of samples drawn from the distribution $\pi$. These samples belong to the subset $D \subseteq X$, where $X = \mathbb{R}^2$. The distribution $\pi$ follows a uniform distribution over the interval $[-2, 2]$.

```python
def sample_uniform_instance_space(n_samples: int,
                                  uni_lower_bound: float,
                                  uni_upper_bound: float) -> np.ndarray:
    """
    Generate samples in R^2 from 2 independent uniform distribution with the same support.
    
    input:
    ------
    n_samples: int represents the number of isntances we want to generate
    uni_lower_bound: float the lower bound of the uniform distribution support
    uni_upper_bound: float the upper bound of the uniform distribution support
    
    return
    ------
    samples: 2d numpy array containing samples distributed according to 2 indepndent uniform distributions
    """
    samples = np.random.uniform(low=uni_lower_bound, high=uni_upper_bound, size=(n_samples, 2))

    return samples
```

# Sample complexity

In class, we discussed the sample complexity bound for the case of concentric circles. Given $\varepsilon$, the desired maximal true error, and $1 - \delta$, the probability of achieving this error, the sample size $m(\varepsilon, \delta)\in O(f(\varepsilon), g(\delta))$. Use this bound to determine the minimum number of samples required to achieve these guarantees.

```python
def sample_bound(true_error: float, uncertainty: float) -> int:
    """
    Return the sample complexity of cocentric circle in R^2 problem
    
    input:
    ------
    true_error: float desired upper bound on our true error
    uncertainty: float 1 - probability to get this error for a random dataset
    
    return:
    -------
    m: int the number of minimal samples
    """
    return int(np.ceil((np.log(1/uncertainty) + np.log((r_concept**2) * np.pi)) / true_error))
```

# Consistent learner

```python
def learning_algorithm_1(D: np.ndarray, c: np.ndarray) -> float:
    """
    The concentric circles consistent learning algorithm
    
    input:
    ------
    D: a 2d numpy array consist of the training instances
    c: a 1d numpy array consist of the concept values (label) 
    
    return:
    -------
    hypothesis_r: float the radius of our learned hypothesis
    """
    hypothesis_r = 1e-6
    is_instance_belongs_to_concept = c == 1

    # Handle the case: in our dataset all the labels are the same
    if is_instance_belongs_to_concept.sum() == 0:
        return hypothesis_r
   
    for i in range(len(c)):
        if c[i]:
            euc_d = (D[i][0] ** 2) + (D[i][1] ** 2)
            hypothesis_r = max(euc_d, hypothesis_r)
        
    
    return hypothesis_r

def learning_algorithm(D: np.ndarray, c: np.ndarray) -> float:
  """
  The concentric circles consistent learning algorithm
  
  input:
  ------
  D: a 2d numpy array consist of the training instances
  c: a 1d numpy array consist of the concept values (label) 
  
  return:
  -------
  hypothesis_r: float the radius of our learned hypothesis
  """
  hypothesis_r = 1e-6
  is_instance_belongs_to_concept = c == 1
  max_1_d = 0
  min_0_d = concept_r * 2 # arbitrarily large

  # Handle the case: in our dataset all the labels are the same
  if is_instance_belongs_to_concept.sum() == 0:
    return hypothesis_r

  for i in range(len(c)):
      euc_d = (D[i][0] ** 2) + (D[i][1] ** 2)
      if c[i]:
        max_1_d = max(euc_d, max_1_d)
      else:
        min_0_d = min(min_0_d, euc_d)


  return (max_1_d + min_0_d) / 2

```

```python
def calculate_true_error(concept_r: float, hypothesis_r: float) -> float:
    """
    Calculating the true error of concentric circle under the uniform distribution over [-2, 2]
    
    input:
    ------
    concept_r: float the radius of the concept 
    hypothesis_r: float the radius of the learnt hypothesis
    
    return:
    error_rate: float the true error
    """
    return abs((np.pi*(concept_r ** 2)) - (np.pi*(hypothesis_r ** 2))) / 16
```

```python
def run_single_experiment(n_samples: int, concept_r: float, epsilon: float, delta: float):
    if n_samples is None:
        raise ValueError(f"n_samples must be a valid integer not None")
        
    # Sample the training dataset
    samples = sample_uniform_instance_space(n_samples=n_samples, uni_lower_bound=uni_lower_bound, uni_upper_bound=uni_upper_bound)
    x1 = samples[:, 0]
    x2 = samples[:, 1]
    c_train = concept_func(x1, x2)
    # This is the actual experiment data set (D)
    train_data = np.vstack([x1, x2, c_train]).T
    hypothesis_r = learning_algorithm(D=train_data[:, :-1], c=train_data[:, -1])
    error_rate = calculate_true_error(concept_r=concept_r, hypothesis_r=hypothesis_r)
    
    return error_rate

def plot_many_experiments(n_samples: int, n_exp: int, concept_r: float, epsilon: float, delta: float):
    error_rates = np.array([run_single_experiment(n_samples=n_samples, concept_r=concept_r, epsilon=epsilon, delta=delta) for _ in range(n_exp)])
    
    # calc empirical delta and average error rate
    empirical_delta = np.mean(error_rates > epsilon)
    average_error_rate = np.mean(error_rates)
    
    sample_complexity = sample_bound(true_error=epsilon, uncertainty=delta)
    
    insufficient_experiments = np.argwhere(error_rates > epsilon).flatten()
    sufficient_experiments = np.argwhere(error_rates <= epsilon).flatten()
    
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.scatter(sufficient_experiments, error_rates[sufficient_experiments], s=1, c='green', label=f"experiments with π(h=L(D),c) ≤ {epsilon} ({len(sufficient_experiments)} experiments)")
    ax.scatter(insufficient_experiments, error_rates[insufficient_experiments], s=1, c='red', label=f"experiments with π(h=L(D),c) > {epsilon} ({len(insufficient_experiments)} experiments)")
    ax.axhline(y=epsilon, color='blue', linestyle='-', label='desired ε')
    
    ax.set_xlim(0, n_exp)
    ax.set_ylim(0, max(error_rates) * 1.1)
    
    ax.set_xlabel("Dataset Index")
    ax.set_ylabel("error π(h=L(D),c)")
    ax.set_title(f"Sample Complexity: {sample_complexity}\nEmpirical 1 - δ: {1 - empirical_delta:.4f}\nAverage Error π(h=L(D),c): {average_error_rate:.4f}")
    
    ax.legend(loc='upper right')
    
    textstr = f"Sample complexity: {sample_complexity}\nEmpirical δ: {empirical_delta:.4f}\nAverage error π(h=L(D),c): {average_error_rate:.4f}"
    props = dict(boxstyle='round', facecolor='white', alpha=0.5)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10, verticalalignment='top', bbox=props)
    
    plt.show()
    
```

```python
epsilon = 0.02
delta = 0.05
n_samples = sample_bound(true_error=epsilon, uncertainty=delta)

plot_many_experiments(n_samples=n_samples, n_exp=n_exp, concept_r=concept_r, epsilon=epsilon, delta=delta)
```
